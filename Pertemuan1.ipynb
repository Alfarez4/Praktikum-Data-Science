## If statement

```python
age = 8
if age >= 18:
    print("anda Dewasa")
elif age >= 13:
    print("Anda Remaja")
else:
    print("Anda anak-anak")
```

```python
countries = ['Indonesia', 'Russia', 'Tiongkok', 'Korea Utara']

if 'Tiongkok' in countries:
    print("Negara dalam list")
else:
    print("Negara tidak dalam list")
```

## For Looping (Perulangan)

```python
for country in countries:
    print(country)
```

```python
for negara in countries:
    print(negara)
    
print("=========")

for negara in countries:
    if negara == 'Indonesia':
        print(negara)
    
print("=========")

for i, negara in enumerate(countries):
    print(i, negara)
```
